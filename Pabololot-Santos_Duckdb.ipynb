```python
#Ensure 'Order ID' is string temporarily, then filter rows with only digits
final_df = final_df[final_df['Order ID'].astype(str).str.isdigit()].copy()

#Strip string columns
string_cols = ["Product", "Order Date", "Purchase Address"]
for col in string_cols:
    final_df.loc[:, col] = final_df[col].astype(str).str.strip()

#Convert numeric fields AFTER cleaning
final_df.loc[:, "Order ID"] = final_df["Order ID"].astype(int)
final_df.loc[:, "Quantity Ordered"] = final_df["Quantity Ordered"].astype(int)
final_df.loc[:, "Price Each"] = final_df["Price Each"].astype(float)
```
